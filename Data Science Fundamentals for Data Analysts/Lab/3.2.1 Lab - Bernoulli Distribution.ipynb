## Exercise 1
Random variable X represents whether we roll a number divisible by 3 on a six sided die.
- Let x = 1 represent success (rolling a 3 or 6)
- Let x = 0 represent failure (rolling a 1, 2, 4, or 5)

The probability of an outcome n can be computed using the probability density function of a Bernoulli distribution:

P(x) = p<sup>x</sup>(1-p)<sup>1-x</sup>, where p is the probability of success.

Compute the probability of success (x=1) using this function.

```
# TODO
p = 2/6
x = 1
prob_success = p**x * (1-p)**(1-x)
print(f"The probability of success is {prob_success}")
```

## Exercise 2

Use the same function to compute the probability of failure (x=0).

```
# TODO
prob_failure = 1 - prob_success
print(f"The probability of failure is {prob_failure}")

# ANSWER (SOLUTION)
p = 1/3
x = 0
prob_failure = p**x * (1-p)**(1-x)
print(f"The probability of failure is {prob_failure}")
```

## Exercise 3

Get the expected value and variance of this distribution.
- expected value = p
- variance = p(1-p)

```
# TODO
p = 2/6
expected_value = p
variance = p * (1 - p)

print(f"The expected value is {expected_value} and the variance is {variance}")
```

## Exercise 4

So far, we've been working with probabilities using Python as a calculator. However, we can also work with probabilities using the `scipy.stats` module.

Complete the below code block to compute the expected value (mean) and varience using `scipy.stats`.

<img alt="Hint" title="Hint" style="vertical-align: text-bottom; position: relative; height:1.75em; top:0.3em" src="https://files.training.databricks.com/static/images/icon-light-bulb.svg"/>&nbsp;**Hint:** You can use the [documentation](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.bernoulli.html) as a reference.

```
# ANSWER (SOLUTION)
from scipy.stats import bernoulli
expected_value, variance = bernoulli.stats(p, moments="mv")
print(f"The expected value is {expected_value} and the variance is {variance}")
```
